## 3. ***Data Wrangling***

### Data Wrangling Code

```python
#Droping subset of names and host name will make no diffrence as they are negligible percentage but we can not drop subset of reviews and reviews per month
#columns because they are 20% of data.Therefore we havw to delete whole column.
```

```python
man_airbnb_df=airbnb_df.drop(["last_review","reviews_per_month"],axis=1)                                                  #df1 is a dataframe which has no null values review column
man_airbnb_df=man_airbnb_df.dropna(subset=["name","host_name"],axis=0)
```

```python
man_airbnb_df.isna().sum()
```

### What all manipulations have you done and insights you found?

Manipulations done in the original dataset i.e airbnb_df are mentiond below:


1)Droped corresponding subset of the null values of columns named "names" and "host_names" which will make no impact in the originality of data.



2)Removed whole columns named "last_review" and "reviews_per_month" because droping the corresponding values will dilute the orignality of data and the insights will be mis-leading as a big chunk of data will be removed.


After doing required manipulations,I have assigned a varibale "man_airbnb_df" to the manipulated dataset that stands for manipulated airbnb dataframe.




## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

Types of rooms people are prefering.

```python
room_pre_df=man_airbnb_df.groupby("room_type").agg({"number_of_reviews":"sum"}).reset_index()
```

```python
room_pre_df
```

```python
colours=["#fd7f6f","#1984c5","#8bd3c7"]
plt.bar(room_pre_df["room_type"],room_pre_df["number_of_reviews"],width=0.3,color=colours)
plt.xlabel("Room Types")
plt.ylabel("Total Bookings")
plt.title("Bar Graph showing room type preferrence")
```

##### 1. Why did you pick the specific chart?

For visualising the room preference, I have used bar graph because they are:

-Easily understandable.

-Easy to compare the catogrical data.




##### 2. What is/are the insight(s) found from the chart?

Entire home/apt are most prefered room type followed by private rooms and shared rooms are least prefered.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The gained insight will defenitly help creating positive impact on business if the reason behind the stunted growth of shared room are known and addressed properly.

#### Chart - 2

Busy or more prefered neighbourhood.

```python
nebourhood_group_df=man_airbnb_df.groupby("neighbourhood_group").agg({"number_of_reviews":"sum"}).reset_index()
```

```python
nebourhood_group_df
```

```python
plt.figure(figsize=(8,8))
plt.pie(nebourhood_group_df["number_of_reviews"],labels=nebourhood_group_df["neighbourhood_group"],autopct="%1.2f%%",)
plt.title("Preferred Neighbourhood")
```

##### 1. Why did you pick the specific chart?

Here for communicating insights on the neighbourhood group preference,I used pie chart because pie chart are:

-Commonly use to represent the relative portion os a set of values or to compare parts of whole.

-Easy and simple to identify the largest and smallest portion.


##### 2. What is/are the insight(s) found from the chart?

Brooklyn is the most prefered or busy neighbourhood group followed by Manhatten,though the diffrence between them is not much.After them there is Queens then Bronx and Staten Island at the last
